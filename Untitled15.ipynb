```python
class Dog():
    """A simple attempt to model a dog"""
    def __init__(self,name,age):  #special method #constructor
        self.name=name #initiailize the name and age attributes
        self.age=age
    def sit(self):
        """dog will response to command"""
        print(f"{self.name} is now sitting")
    def roll_over(self):
        """dog will rollover to command"""
        print(f"{self.name} now rollover")
my_dog=Dog('wille',5)  
print(f"my dog's name is {my_dog.name}")
print(f"my dog is {my_dog.age}years old")
my_dog.sit()
my_dog.roll_over()
```
