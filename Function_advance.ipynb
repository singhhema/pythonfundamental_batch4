```python
def greet_user():  #Define a Function
    """Display a simple greeting"""  #Doc-string---
    print("Hello")
greet_user()   #calling a function
```

```python
def greet_user(username):   #parameter passing
    """Display a simple greeting to user"""
    print(f"Hello,{username.title()}!")
greet_user('abhi')  # argument passing
```

```python
greet_user('hema s')
```

```python
#Type of arguments:
#positional arguments
#keywork arguments
def describe_pet(animal_type,pet_name):
    """Display information about a pet"""
    print(f"\nI have a {animal_type}")
    print(f"my {animal_type}'s name is {pet_name}")
describe_pet('dog','sheru')
```
