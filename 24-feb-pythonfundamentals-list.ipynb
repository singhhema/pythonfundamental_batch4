```python
#How to define a list

bicycle=['ranger','hero','hercules']
print(bicycle)
```

```python
#Access elements in a list
print(bicycle[1])  #to access element 2
```

```python
print(bicycle[-2]) #access element two from reverse end
```

```python
print(bicycle[0].title)
```

```python
print(bicycle[0].title())
```

```python
message=f"my first bicycle was a (bicycle[0].title)"
```

```python
print(message)
```

```python
message=f"my first bicycle was a {bicycle[0].title}"
print(message)
```

```python
message=f"my first bicycle was a {bicycle[0].title()}"
print(message)
```
